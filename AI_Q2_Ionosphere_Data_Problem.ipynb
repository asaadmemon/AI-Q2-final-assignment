```python
ionosphere_data = pd.read_csv('ionosphere_data.csv')
ionosphere_data.head()
```

```python
ionosphere_data.isnull().sum()
```

```python
from sklearn.utils import shuffle
ionosphere_data = shuffle(ionosphere_data)
ionosphere_data.head()
```

```python
ionosphere_data.shape
```

```python
train_data = ionosphere_data.iloc[0:210, 0:-1]
train_targets = ionosphere_data.iloc[0:210, -1]

test_data = ionosphere_data.iloc[210:, 0:-1]
test_targets = ionosphere_data.iloc[210:, -1]

print(train_data.shape)
print(train_targets.shape)
print(test_data.shape)
print(test_targets.shape)
```

```python
train_targets.head()
```

```python
ionosphere_data['feature1']
```

```python
ionosphere_data['feature2']
```

```python
train_data.drop('feature2', inplace=True, axis=1)
```

```python
test_data.drop('feature2', inplace=True, axis=1)
```

```python
mean = train_data.mean()
std = train_data.std()

train_data-=mean
train_data/=std

test_data-=mean
test_data/=std

train_data.head()
```

```python
test_data.head()
```

```python
train_targets = train_targets.map({'g':1, 'b':0})
train_targets.head()
```

```python
test_targets = test_targets.map({'g':1, 'b':0})
test_targets.head()
```

```python
from keras import models, layers, regularizers

model = models.Sequential()

model.add(layers.Dense(16, input_shape=(train_data.shape[1],), activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(layers.Dense(10, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(layers.Dense(6, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_data, train_targets, validation_split=0.2, epochs=100)
```

```python
model.evaluate(test_data, test_targets)
```

```python
predictions = model.predict(test_data)
len(predictions)
```

```python
import numpy as np
pd.DataFrame({'Actual':test_targets, 'Predicted':np.round(predictions.reshape(141,))})
```
